```python
import logging
import pandas as pd
from simpletransformers.t5 import T5Model, T5Args


logging.basicConfig(level=logging.INFO)
transformers_logger = logging.getLogger("transformers")
transformers_logger.setLevel(logging.WARNING)
```

```python
train_df = pd.read_csv("ass-to-brx/train.tsv", sep="\t").astype(str)
eval_df = pd.read_csv("ass-to-brx/eval.tsv", sep="\t").astype(str)

train_df["prefix"] = ""
eval_df["prefix"] = ""
```

```python
model_args = T5Args()
model_args.max_seq_length = 80
model_args.train_batch_size = 10
model_args.eval_batch_size = 10
model_args.num_train_epochs = 30
model_args.evaluate_during_training = True
model_args.evaluate_during_training_steps = 2000
model_args.use_multiprocessing = False
model_args.fp16 = False
model_args.save_steps = -1
model_args.save_eval_checkpoints = False
model_args.no_cache = True
model_args.reprocess_input_data = True
model_args.overwrite_output_dir = True
model_args.preprocess_inputs = False
model_args.num_return_sequences = 1
model_args.wandb_project = "MT5 assames to bodo Translation"

model = T5Model("mt5", "google/mt5-base", args=model_args)
model.train_model(train_df, eval_data=eval_df)
```

```python
import logging
import sacrebleu
import pandas as pd
from simpletransformers.t5 import T5Model, T5Args


logging.basicConfig(level=logging.INFO)
transformers_logger = logging.getLogger("transformers")
transformers_logger.setLevel(logging.WARNING)


model_args = T5Args()
model_args.max_length = 512
model_args.length_penalty = 1
model_args.num_beams = 10

model = T5Model("mt5", "outputs/best_model", args=model_args)
```

```python
eval_df = pd.read_csv("ass-to-brx/eval.tsv", sep="\t").astype(str)

assames_truth = [eval_df.loc[eval_df["prefix"] == "translate bodo to assames"]["target_text"].tolist()]
to_assames = eval_df.loc[eval_df["prefix"] == "translate bodo to assames"]["input_text"].tolist()

bodo_truth = [eval_df.loc[eval_df["prefix"] == "translate assames to bodo"]["target_text"].tolist()]
to_bodo = eval_df.loc[eval_df["prefix"] == "translate assames to bodo"]["input_text"].tolist()
```

```python
assames_preds = model.predict(to_assames)

brx_ass_bleu = sacrebleu.corpus_bleu(assames_preds, assames_truth)
print("--------------------------")
print("Bodo to assames: ", brx_ass_bleu.score)

bodo_preds = model.predict(to_bodo)

ass_brx_bleu = sacrebleu.corpus_bleu(bodo_preds, bodo_truth)
print("assames to Bodo: ", ass_brx_bleu.score)
```
